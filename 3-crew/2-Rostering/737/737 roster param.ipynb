```python

import pandas as pd
import numpy as np
```

```python
roster_df=pd.read_csv(r'roster757.csv')
crew_df=pd.read_csv(r'Total.csv')
crew_df
```

```python

roster_df
```

```python
# Convert flight information to a dictionary for easy lookup
crew_costs = crew_df.set_index('number')['Total'].to_dict()

# Function to create binary matrix indicating flight coverage by routes
def crew_coverage_matrix(roster):
    coverage = []
    for crew_id in crew_costs.keys():
        if crew_id in roster:
            coverage.append(1)
        else:
            coverage.append(0)
    return coverage

# Create binary matrix for flight coverage by routes
coverage_matrix = roster_df.apply(lambda row: crew_coverage_matrix(row.dropna().tolist()), axis=1)

# Convert the binary matrix to a DataFrame
a = pd.DataFrame(coverage_matrix.to_list(), columns=crew_costs.keys())
a
```

```python
a.to_csv("param.csv")
```

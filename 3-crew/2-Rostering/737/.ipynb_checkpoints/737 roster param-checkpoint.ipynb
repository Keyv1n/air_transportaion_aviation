```python
# Convert roster information to a dictionary for easy lookup
roster_costs = rosters_df.set_index('number')['Total'].to_dict()

# Function to create binary matrix indicating roster coverage by routes
def roster_coverage_matrix(route):
    coverage = []
    for roster_id in roster_costs.keys():
        if roster_id in route:
            coverage.append(1)
        else:
            coverage.append(0)
    return coverage

# Create binary matrix for roster coverage by routes
coverage_matrix = roster_df.apply(lambda row: roster_coverage_matrix(row.dropna().tolist()), axis=1)

# Convert the binary matrix to a DataFrame
a = pd.DataFrame(coverage_matrix.to_list(), columns=roster_costs.keys())
a
```

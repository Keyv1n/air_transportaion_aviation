```python
import pandas as pd
import numpy as np
import csv
import xlsxwriter

df=pd.read_excel(r'fleet asignment result.xlsx')
```

```python
df['o']=df['origin']
df['d']=df['destination']
df['o']=df['o'].replace({'JFK':1, 'LAX':2, 'SFO':3, 'ORD':4, 'ATL':5, 'MIA':6, 'BOS':7, 'IAD':8})
df['d']=df['d'].replace({'JFK':1, 'LAX':2, 'SFO':3, 'ORD':4, 'ATL':5, 'MIA':6, 'BOS':7, 'IAD':8})
```

```python
df
```

```python
#daily combination with min and max duty
a=[]
for i in range(42):
    if df['fleet'].loc[i]==757:
        b=np.zeros(4)
        c=df['flight'].loc[i]
        b[0]=(c)
        a.append(b)
        
    else:
        continue
            
    for j in range(42):                                  #0.16666=10minute  min bitween change a aircraft with coptain                                      #3 =max hour bitween change a aircraft with coptain                                         
        if df['d'].loc[i]==df['o'].loc[j] and (df['arrival'].loc[i] + 0.1666) <= df['departure'].loc[j] and df['departure'].loc[j]<=df['arrival'].loc[i]+3 and df['fleet'].loc[j]==757 :
            b=np.zeros(4)
            c=df['flight'].loc[i]
            d=df['flight'].loc[j]
            b[0]=(c)
            b[1]=(d)

            a.append(b)
           
        else:
            continue
        for z in range(42):
            if df['d'].loc[j]==df['o'].loc[z] and (df['arrival'].loc[j] + 0.1666) <= df['departure'].loc[z] and df['departure'].loc[z]<=df['arrival'].loc[j]+3 and df['fleet'].loc[z]==757 :
                b=np.zeros(4)
                c=df['flight'].loc[i]
                d=df['flight'].loc[j]
                e=df['flight'].loc[z]
                b[0]=(c)
                b[1]=(d)
                b[2]=(e)
                a.append(b)
            
            else:
                continue
            for x in range(42):
                if df['d'].loc[z]==df['o'].loc[x] and (df['arrival'].loc[z] + 0.1666) <= df['departure'].loc[x] and df['departure'].loc[x]<=df['arrival'].loc[z]+3 and df['fleet'].loc[x]==757 :

                    b=np.zeros(4)
                    c=df['flight'].loc[i]
                    d=df['flight'].loc[j]
                    e=df['flight'].loc[z]
                    f=df['flight'].loc[x]
                    b[0]=(c)
                    b[1]=(d)
                    b[2]=(e)
                    b[3]=(f)
                    a.append(b)
                else:
                    continue
aa=a
a=[]  #filter Each duty should not exceed 8 hours of flight time
for i in range(len(aa)):
    h=0
    for j in range (len(aa[i])):
        c=(aa[i][j]-101)
        if c<0:
            continue
        else:
            h=h+df['duration'].loc[c]
    if h<=8:
         a.append(aa[i])
```

```python
#coding each combination invovle how man flight
b=np.zeros(len(a))
for i in range(len(a)):
    for j in range(4):
        if a[i][j]>0:
            b[i]=b[i]+1
for i in range(len(a)):
    b[i]=b[i]-1
len(b)
b
```

```python
#A maximum length of two days is allowed for a routing

c=[]
for i in range(len(a)):   #two days routing     cycle1           cycle2
    c0=a[i][0]                                #[c0,.....c1]    [c2,.....c3]
    f= int(b[i])
    c1=a[i][f]
    for j in range(len(a)):
        c2=a[j][0]
        f=int(b[j])
        c3=a[j][f]
        if df['d'].loc[(c1-101)]==df['o'].loc[(c2-101)]  and   df['d'].loc[(c3-101)]==df['o'].loc[(c0-101)] and df['d'].loc[(c3-101)] ==1 and df['d'].loc[(c1-101)] !=1   :
            d=np.insert(a[i],4,a[j])
            c.append(d)
for i in range(len(a)):   #one days routing
    c0=a[i][0] 
    f= int(b[i])
    c1=a[i][f]
    if df['d'].loc[(c1-101)]==df['o'].loc[(c0-101)] and df['d'].loc[(c1-101)]==1 :
        
        c.append(a[i])
```

```python
file = open('757.csv', 'w+', newline ='') 
with file:     
    write = csv.writer(file) 
    write.writerows(c)
```

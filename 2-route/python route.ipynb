```python
import pandas as pd
import numpy as np
import csv
import xlsxwriter

df=pd.read_excel(r'fleet.xlsx')
df['o']=df['origin']
df['d']=df['destination']
df['o']=df['o'].replace({'IKA':1, 'TJR':2, 'THR':3, 'PSD':4, 'SDQ':5, 'KRJ':6, 'VNK':7 })
df['d']=df['d'].replace({'IKA':1, 'TJR':2, 'THR':3, 'PSD':4, 'SDQ':5, 'KRJ':6, 'VNK':7 })
```

```python
df
```

```python
#hourly cycle loop 
#range(number of flight)


a=[]
for i in range(42):    #range(number of flight) 15 min 
    if df['fleet1'].loc[i]==4:  #name aircarft
        b=np.zeros(3)
        c=df['FlightNum'].loc[i]
        b[0]=(c)
        a.append(b)
        
    else:
        continue
            
    for i1 in range(42):                                             #0.01=10 min bitween arivall andeparture
        if df['d'].loc[i]==df['o'].loc[i1]  and df['fleet1'].loc[i1]==4 and ((df['arrival'].loc[i]+0.007) <= df['departure'].loc[i1]):
            b=np.zeros(3)
            c=df['FlightNum'].loc[i]
            d=df['FlightNum'].loc[i1]
            b[0]=(c)
            b[1]=(d)

            a.append(b)
           
        else:
            continue
        for i2 in range(42):
            if df['d'].loc[i1]==df['o'].loc[i2] and df['fleet1'].loc[i2]==4 and ((df['arrival'].loc[i1]+0.007) <= df['departure'].loc[i2]):
                b=np.zeros(3)
                c=df['FlightNum'].loc[i]
                d=df['FlightNum'].loc[i1]
                e=df['FlightNum'].loc[i2]
                b[0]=(c)
                b[1]=(d)
                b[2]=(e)
                a.append(b)
            
            else:
                continue
            
            
                    
                    
a                  
```
